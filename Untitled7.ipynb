```python
colors_list=["red","purple", "blue", "green", "orange", "yellow"]
```

```python
lado= int (input("Dame el n√∫mero de lados "))
initializeTurtle()
for i in range(lado):#[1,2,3,4]:
  pencolor (colors_list[i % 6])
  width(int (i/100)+1)
  forward(150)
  left(360/ lado)
  for i in range(lado):#[1,2,3,4]:
    pencolor (colors_list[i % 6])
    width(int (i/100)+1)
    forward(100)
    left(360/ lado)
    for i in range(lado):#[1,2,3,4]:
      pencolor (colors_list[i% 6])
      width(int (i/100)+1)
      forward(50)
      left(360/ lado)
```
